```python
# =============================
# MNIST CNN - Train & Save Model
# =============================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
```

```python
# =============================
# 1пёҸвғЈ Load Dataset
# =============================
train = pd.read_csv(r"D:\MNIST_Digit_Recognition_Project\Dataset\train.csv")
test  = pd.read_csv(r"D:\MNIST_Digit_Recognition_Project\Dataset\test.csv")
```

```python
# =============================
# 2пёҸвғЈ Split features and labels
# =============================
x_train = train.iloc[:, 1:].values
y_train = train.iloc[:, 0].values

x_test = test.values  # Test dataset (labels not included)
```

```python
# =============================
# 3пёҸвғЈ Normalize & Reshape for CNN
# =============================
x_train = x_train / 255.0
x_test  = x_test / 255.0

x_train = x_train.reshape(-1, 28, 28, 1)
x_test  = x_test.reshape(-1, 28, 28, 1)

y_train = to_categorical(y_train, num_classes=10)
```

```python
# =============================
# 4пёҸвғЈ Train-Validation Split
# =============================
x_train_part, x_val, y_train_part, y_val = train_test_split(
    x_train, y_train, test_size=0.1, random_state=42
)
```

```python
# =============================
# 5пёҸвғЈ Build CNN Model
# =============================
model = Sequential([
    Conv2D(32, kernel_size=(3,3), activation='relu', input_shape=(28,28,1)),
    MaxPooling2D(pool_size=(2,2)),
    
    Conv2D(64, kernel_size=(3,3), activation='relu'),
    MaxPooling2D(pool_size=(2,2)),
    
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.3),
    Dense(10, activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
# =============================
# 6пёҸвғЈ Train the Model
# =============================
history = model.fit(
    x_train_part, y_train_part,
    validation_data=(x_val, y_val),
    epochs=10,
    batch_size=64
)
```

```python
# =============================
# 7пёҸвғЈ Save the Model
# =============================
model.save(r"D:\MNIST_Digit_Recognition_Project\mnist_cnn.h5")
print("Model saved at: D:\\MNIST_Digit_Recognition_Project\\mnist_cnn.h5")
```
